```python
# %% Create a max pooling instance

# Parameters
size   = 3
stride = 3

# Class instances
p2 = nn.MaxPool2d(size,stride)
p3 = nn.MaxPool3d(size,stride)

print(p2)
print(p3)
```

```python
# %% generate some data and apply max pooling

img_2d = torch.randn(1,1,30,30)
img_3d = torch.randn(1,3,30,30)

# all combinations of image and maxpool dimensionality (pool 3D with a 2D input
# is not possible of course)
img2_pool2 = p2(img_2d)
print(f'2D image, 2D maxpool: {img2_pool2.shape}\n' )

img3_pool2 = p2(img_3d)
print(f'3D image, 2D maxpool: {img3_pool2.shape}\n' )

img3_pool3 = p3(img_3d)
print(f'3D image, 3D maxpool: {img3_pool3.shape}\n' )

```

```python
# Create a simple CNN

little_cnn = nn.Sequential(

    # Convolution-pooling block
    nn.Conv2d(3,10,5,3,2),   # convolution layer
    nn.ReLU(),             # activation function
    nn.AvgPool3d(3,3),     # average-pool

    # FFN block
    nn.Flatten(),          # vectorise image
    nn.Linear(588,1),      # fully connected linear layer
    nn.Sigmoid()           # output activation

)
```

```python
# %% Test with a bit of data

img = torch.randn(1,3,128,128)
little_cnn(img)
```
