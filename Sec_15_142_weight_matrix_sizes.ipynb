```python
# %% Build a toy model

model_a = nn.Sequential( nn.Linear(10,15),
                         nn.Linear(15,21),
                         nn.Linear(21,7)
                         )

model_a
```

```python
# %% Inspect its weight matrices sizes

for i in range(len(model_a)):
    print(model_a[i].weight.shape)
```

```python
# %% Build a layer with inconsistent layer shapes

model_b = nn.Sequential( nn.Linear(10,15),
                         nn.Linear(11,25),
                         nn.Linear(21,7)
                         )

for i in range(len(model_b)):
    print(model_b[i].weight.shape)
```
